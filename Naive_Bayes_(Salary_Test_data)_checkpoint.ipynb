```python
# Splitting the data into x and y
x = df.drop(["Salary"], axis=1)
y = df["Salary"]
```

```python
# splitting x and y into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=0)
```

```python
# display categorical variables
cat = [col for col in x_train.columns if x_train[col].dtype=='O']
cat
```

```python
# display numerical variables
num = [col for col in x_train.columns if x_train[col].dtype != 'O']
num
```

# Data Transformation

```python
import category_encoders as ce
encoder = ce.OneHotEncoder(cols=["workclass","education","maritalstatus","occupation","relationship","race","sex","native"])

x_train = encoder.fit_transform(x_train)
x_test = encoder.fit_transform(x_test)
x_train.head()
```

```python
cols = x_train.columns
```

```python
# importing StandrdScaler
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()

x_train = SS.fit_transform(x_train)
x_test = SS.fit_transform(x_test)

x_train = pd.DataFrame(x_train, columns=[cols])
x_test = pd.DataFrame(x_test, columns=[cols])
x_train.head()
```

```python
# Model fitting
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()

from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
# fit the model
GNB.fit(x_train, y_train)

# prediction
y_pred_train = GNB.predict(x_train)
y_pred_test = GNB.predict(x_test)
```

```python
# accuracay
from sklearn.metrics import accuracy_score
ac1 = accuracy_score(y_train,y_pred_train)
print("training score:", ac1.round(2))
ac2 = accuracy_score(y_test,y_pred_test)
print("test score:", ac2.round(2))
```

```python
# Bernoulli's
from sklearn.naive_bayes import BernoulliNB
BNB = BernoulliNB()
#fit the model
BNB.fit(x_train,y_train)
# prediction
y_pred_train = BNB.predict(x_train)
y_pred_test = BNB.predict(x_test)
```

```python
# accuracy score
from sklearn.metrics import accuracy_score
ac1 = accuracy_score(y_train,y_pred_train)
print("training score:", ac1.round(2))
ac2 = accuracy_score(y_test,y_pred_test)
print("test score:", ac2.round(2))
```

```python
# SVC
from sklearn.svm import SVC
svc_class = SVC(kernel='linear')
#fit the model
svc_class.fit(x_train,y_train)
# prediction
y_pred_train = svc_class.predict(x_train)
y_pred_test = svc_class.predict(x_test)
```

```python
# accuracy score
from sklearn.metrics import accuracy_score
ac1 = accuracy_score(y_train,y_pred_train)
print("training score:", ac1.round(2))
ac2 = accuracy_score(y_test,y_pred_test)
print("test score:", ac2.round(2))
```

```python
# LogisticRegression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
# fit the model
logreg.fit(x_train,y_train)
# prediction
y_pred_train = logreg.predict(x_train)
y_pred_test = logreg.predict(x_test)
```

```python
# accuracy score
ac1 = accuracy_score(y_train,y_pred_train)
print("training score:", ac1.round(2))
ac2 = accuracy_score(y_test,y_pred_test)
print("test score:", ac2.round(2))
```
